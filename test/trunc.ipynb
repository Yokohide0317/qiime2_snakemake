```python
import pandas as pd
```

```python
df = pd.read_csv("../02_adapter/out/forward-seven-number-summaries.tsv", sep="\t", index_col=0)
df
```

```python
# 最後の方はちょっとしかリードが残っていないので、数パーセント以下のところを削除する。
CUT_COUNT_PERCENT: float = 0.9

def cut_edge(_df: pd.DataFrame, _CUT_COUNT_PERCENT: float):
    max_count: int = int(_df.loc["count"][0])

    drop_colnum: int = int(_df.columns[-1])
    for col, data in enumerate(_df.loc["count"]):
        if data < max_count * _CUT_COUNT_PERCENT:
            drop_colnum = int(col)
            break

    filt_df = _df.drop([str(col) for col in _df.columns[drop_colnum:]], axis=1)
    return filt_df

filt_df = cut_edge(df, CUT_COUNT_PERCENT)
```

```python
# 25%のQualityが20%を切るタイミングを探す
CUT_QUALITY_PERCENT = 20
CUT_QUALITY_LOCATION = 25

def quality_check(_df: pd.DataFrame, _CUT_QUALITY_PERCENT: int, _CUT_QUALITY_LOCATION: int) -> int:
    target_row = _df.loc[f"{str(_CUT_QUALITY_LOCATION)}%"]

    trunc_loc: int = int(target_row.keys()[-1])
    for col, data in target_row.items():
        if data < _CUT_QUALITY_PERCENT:
            trunc_loc = int(col) -1
            break

    return trunc_loc

quality_check(filt_df, CUT_QUALITY_PERCENT, CUT_QUALITY_LOCATION)
```

```python
CUT_COUNT_PERCENT: float = 0.9
CUT_QUALITY_PERCENT = 25
CUT_QUALITY_LOCATION = 50

# FWD
tsv_path = "../02_adapter/out/forward-seven-number-summaries.tsv"
trunc_len_f = quality_check(cut_edge(pd.read_csv(tsv_path, sep="\t", index_col=0), CUT_COUNT_PERCENT), CUT_QUALITY_PERCENT, CUT_QUALITY_LOCATION)

# REV
tsv_path = "../02_adapter/out/reverse-seven-number-summaries.tsv"
trunc_len_r = quality_check(cut_edge(pd.read_csv(tsv_path, sep="\t", index_col=0), CUT_COUNT_PERCENT), CUT_QUALITY_PERCENT, CUT_QUALITY_LOCATION)

print(trunc_len_f)
print(trunc_len_r)
```
